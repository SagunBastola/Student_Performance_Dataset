```python
# validation
student_data=student_data[round((student_data["Math_Score"]+student_data["English_Score"]+student_data["Science_Score"])/3,ndigits=2)==student_data["Final_Percentage"]]
```

```python
student_data.groupby("Performance_Level")["Final_Percentage"].min()
```

```python
student_data.groupby("Performance_Level")["Final_Percentage"].max()
```

```python
label=["Should Improve","Better","Very Good","Hardworkig"]
bins=[0,50,70,85,100]
student_data["Remarks"]=pd.cut(student_data["Final_Percentage"],labels=label,bins=bins)
```
